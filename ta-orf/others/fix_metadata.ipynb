```python
import numpy as np
import pandas as pd
import os
```

```python
plate_maps = pd.read_csv("TAORF_REFERENCE_SET_2.txt", sep="\t")
plate_maps.columns = ['Metadata_Well', 'gene_name', 'pert_name', 'broad_sample', 'cell_line', 'ASSAY_WELL_ROLE', 'GeneID', 'plate_map_name']
plate_maps
```

```python
index = pd.read_csv("/raid/data/cellpainting/TAORF/inputs/metadata/index.csv")
index = index.drop(["gene_name", "Treatment", "gene_name_Replicate", "val"], axis=1)
index
```

```python
new_idx = pd.merge(index, plate_maps, on="Metadata_Well")
new_idx
```

```python
new_idx["pert_name_replicate"] = 0
for pert in new_idx.pert_name.unique():
    view = new_idx[new_idx.pert_name == pert].groupby(["Metadata_Plate", "Metadata_Well"])["DNA"].count().reset_index()
    replicate = 1
    for k,r in view.iterrows():
        mask1 = new_idx.Metadata_Plate == r.Metadata_Plate
        mask2 = new_idx.Metadata_Well == r.Metadata_Well
        new_idx.loc[mask1 & mask2, "pert_name_replicate"] = replicate
        replicate += 1
```

```python
new_idx
```

```python
new_idx["val"] = 0
new_idx.loc[new_idx["pert_name_replicate"] > 3, "val"] = 1
new_idx.loc[new_idx["pert_name_replicate"] > 97, "val"] = 0
```
